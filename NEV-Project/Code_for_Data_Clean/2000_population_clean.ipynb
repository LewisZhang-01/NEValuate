# Introduction

In this NoteBook we will clean the dataset of population of each state.

## Set up

```python
import pandas as pd
```

## Read the data

```python
population = pd.read_csv('../Raw_Data/US_population.csv', skiprows=[*range(0, 4)],nrows=52, delimiter = ',')
population.head()
```

We only keep 1950-2021 data 

```python
population.set_index('GeoName',inplace=True)
# we don't need column "GeoFips" , so clean this column
population = population.iloc[1: ,22:]
population.rename(index={'Alaska *':'Alaska','Hawaii *':'Hawaii'},inplace=True)
print(population.shape)
population.head()
```

```python
path = '../Processed_Data/population.csv'
population.to_csv(path)
```
