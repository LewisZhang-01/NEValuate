# Introduction

In this Notebook, we will clean the data of personal income of different states

## Set up

```python
import pandas as pd
```

## Read the Data

```python
#read the data file, skip the title and note 
data = pd.read_csv('../Raw_Data/Per_capita_personal_income.csv',skiprows=[*range(0, 4)],nrows=52, delimiter = ',')
```

```python
data.head()
```

## Data clean

```python
data.set_index('GeoName',inplace=True)
```

```python
# we don't need column "GeoFips" , so clean this column until 1980
data = data.iloc[1: ,52:]
```

```python
data.head()
```

```python
data.rename(index={'Alaska *':'Alaska','Hawaii *':'Hawaii'},inplace=True)
print(data.shape)
data.head()
```

```python
#save the cleaned data to processed file
path = '../Processed_Data/incomes.csv'
data.to_csv(path)
```
