# Introduction

In this NoteBook, we will clean the dataset of GDP for each states

## Set up

```python
import pandas as pd
```

## Read Data

```python
# read the file and skip the title and note at the end 
GDP = pd.read_csv('../Raw_Data/annual_GDP_by_state.csv', skiprows=[*range(0, 4)],nrows=52, delimiter = ',')
```

```python
GDP.head()
```

```python
GDP.set_index('GeoName',inplace=True)
# we don't need column "GeoFips" , so clean this column
GDP = GDP.iloc[1: ,1:]
```

```python
print(GDP.shape)
GDP.head()
```

**Because this is a anual GDP per state, we need per capita personal GDP**

```python
population = pd.read_csv('../Processed_Data/population.csv', delimiter = ',', index_col=0)
population.head()
```

**we need 1997-2021 year population data**

```python
population = population.iloc[:,47:]
print(population.shape)
population.head()
```

```python
GDP_per_capita = GDP/population
GDP_per_capita.head()
```

**the value are in million, we change to dollar**

```python
GDP_per_capita = (GDP_per_capita*1000000).round(decimals = 2)
GDP_per_capita.head()
```

```python
path = '../Processed_Data/gdp.csv'
GDP_per_capita.to_csv(path)
```
