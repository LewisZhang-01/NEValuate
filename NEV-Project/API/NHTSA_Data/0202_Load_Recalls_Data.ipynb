# Load and aggregate Recalls Data

```python
import json
from glob import glob
import pandas as pd
import numpy as np
import itertools

from loguru import logger
from IPython.display import display, clear_output
```

Path to recalls dataset

```python
complaints_dataset = "../raw/Recalls"
```

Aggregate all json file in this directory.

```python
files = glob('../raw/Recalls/*.json')
len(files)
```

```python
data = []

for i, file in enumerate(files):
    
    clear_output()
    logger.info((i, file))
    
    with open(file, 'r') as f: data.append(json.load(f))
        
df = pd.DataFrame(itertools.chain.from_iterable(data))
df.shape
```
